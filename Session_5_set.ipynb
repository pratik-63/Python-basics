# 5) set oprations

```python
# Union
set1 = {1,2,3,4}
set2 = {3,4,5,6}
result = set1 | set2
print (result)
```

```python
result2 = set1.union(set2)
print(result2)
```

```python
set3 = {7,8,9}
result = set1 | set2 | set3
print (result)
```

```python
# intersection 
set1 = {1,2,3,4}
set2 = {3,4,5,6}
result = set1 & set2 
print (result)
```

```python
result = set1.intersection(set2)
print (result)
```

```python
# difference
set1 = {1,2,3,4}
set2 = {8,9,5,6}
result = set2 -set1
print (result)
```

```python
result = set1.difference(set2)
print(result)
```

```python
# symmetric difference
set1 = {1,2,3,4}
set2 = {3,4,5,6}
result = set1 ^ set2
print (result)
```

```python
result = set1.symmetric_difference(set2)
print(result)
```

# 8) set comprehension 

```python
# traditional way
squares = set ()
for x in range (1,6):
    squares.add(x**2)
print (squares)
```

```python
squares = {x**2 for x in range (1,6)}
print (squares)
```

```python
# print even numbers only
evens = {x for x in range (1,11) if x % 2 ==0}
print (evens)
```
