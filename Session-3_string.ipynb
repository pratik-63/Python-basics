# 1) breake , continue , pass 

```python
# breake statement
for i in range(1,11):    # break immediately stops the loop
    if i==6:
        break
    print(i)
```

```python
# continue statement
for i in range(1,11):
    if i % 2 == 0:    #current value skip karto 
        continue # Skip Number # two number cha jump aahe 
    print(i)
```

```python
# pass statement
for i in range(5):     # pass dose nothing it is used as a placeholder for future code  [pasa kahich kam karat nahi ]
    if i==2:
        pass             # Do Nothing
    print(i)
```
