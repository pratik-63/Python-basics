# 11) list comprehension
concised way to create list using line od code


```python
#  syntax: [expression for item in iterabe in condition]
square =[]
for x in range (1,6):
    square . append (x * x)
print (square)
```

```python
# single line code 

square =[x ** 2 for x in range (1,6)]
print (square)
```

```python
cube = [x ** 3 for x in range (1,6)]
print (cube)
```

```python
evens = [x for x in range (1 , 11) if x % 2 ==0]
print (evens)
```

```python
divideby3 = [x for x in range (1,21)if x % 3 == 0]
print (divideby3)
```
