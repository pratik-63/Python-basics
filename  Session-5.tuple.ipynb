# 11) zip function :
combines multiple iterable into tuple

```python
names = ("pratik","sohan","chetan")
ages = (25 ,22,28)
combined = tuple(zip(names,ages))
print (combined)
```

```python
# unzip 
combinrd =  (('pratik',23),('sohan',22),('chetan',28))
names ,ages = zip (*combinrd)
print (names)
print (ages)
```

```python
tuple1 = (1,2,3,4)
tuple2 = ("a","b")
result = tuple(zip(tuple1, tuple2))
print (result)
```
